```python
import torch
```

```python
X = torch.tensor(
    [
        [0,0],
        [0,1], 
        [1,0], 
        [1,1]
    ],  dtype = torch.float 
)
Y = torch.tensor(
    [
        [0],
        [1],
        [1],
        [0]
    ],  dtype = torch.float
)
print(X.shape, Y.shape)
```

```python
class XORNet(torch.nn.Module):
    def __init__(self) -> None:
        super().__init__()
        self.layer_1 = torch.nn.Linear(2, 2)
        self.non_li1 = torch.nn.LeakyReLU(negative_slope = 0.1)
        self.layer_2 = torch.nn.Linear(2, 1)
        self.print_mode = False
    
    def print_out(self, x : torch.Tensor):
        if self.print_mode:
            print(x)

    def forward(self, x : torch.Tensor) -> torch.Tensor:
        x = self.layer_1(x)
        self.print_out(x)
        x = self.non_li1(x)
        self.print_out(x)
        x = self.layer_2(x)
        self.print_out(x)
        return x
```

```python
model = XORNet()
lossF = torch.nn.MSELoss()
optim = torch.optim.Adam(model.parameters(), lr = 0.01)
```

```python
for i in range(1_000):
    xi = torch.unsqueeze(X, dim = 0)
    yi = torch.unsqueeze(Y, dim = 0)

    optim.zero_grad()
    yp = model(xi)
    ls = lossF(yp, yi)
    ls.backward()
    optim.step()
    
    if i % 100 == 0:
        print(i, ls.item())
```

```python
model.print_mode = True
```
